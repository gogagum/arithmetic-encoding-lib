```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = open("../build/benchmark/results.json")
data = json.load(file)
```

```python
benchmarks = pd.DataFrame(data['benchmarks'])
```

```python
benchmarks
```

```python
benchmarks[["coder", "input_size", "h_quater"]]\
  = benchmarks["name"].str.split('/', expand=True)
```

```python
benchmarks = benchmarks.astype({"input_size" : "int64"})
```

```python
benchmarks
```

```python
set(benchmarks["coder"].to_list())
```

```python
def getBenchmark(coder_name, h_quater):
    return benchmarks.where(benchmarks.coder == coder_name).where(benchmarks.h_quater == h_quater).dropna()
```

```python
getBenchmark("BM_benchmark_adaptive_coder", "2")
```

```python
fig, ax = plt.subplots(3, 3, figsize=(15, 15))

data_adaptive = [getBenchmark("BM_benchmark_adaptive_coder", i) for i in ["0", "1", "2", "3", "4"]]

input_size_adaptive = [data_adaptive[i]['input_size'] for i in range(5)]
real_time_adaptive = [data_adaptive[i]['real_time'] for i in range(5)]
speed_adaptive = [real_time_adaptive[i] / input_size_adaptive[i] for i in range(5)]

for i in range(5):
    ax[0][0].plot(input_size_adaptive[i], speed_adaptive[i], label=f"h_quater={i}")

ax[0][0].set_yscale("log")
ax[0][0].legend()

data_adaptive_a = [getBenchmark("BM_benchmark_adaptive_a_coder", i) for i in ["0", "1", "2", "3", "4"]]

input_size_adaptive_a = [data_adaptive_a[i]['input_size'] for i in range(5)]
real_time_adaptive_a = [data_adaptive_a[i]['real_time'] for i in range(5)]
speed_adaptive_a = [real_time_adaptive_a[i] / input_size_adaptive_a[i] for i in range(5)]

for i in range(5):
    ax[0][1].plot(input_size_adaptive_a[i], speed_adaptive_a[i], label=f"h_quater={i}")

ax[0][1].set_yscale("log")
ax[0][1].legend()

data_adaptive_d = [getBenchmark("BM_benchmark_adaptive_d_coder", i) for i in ["0", "1", "2", "3", "4"]]

input_size_adaptive_d = [data_adaptive_d[i]['input_size'] for i in range(5)]
real_time_adaptive_d = [data_adaptive_d[i]['real_time'] for i in range(5)]
speed_adaptive_d = [real_time_adaptive_d[i] / input_size_adaptive_d[i] for i in range(5)]

for i in range(5):
    ax[0][2].plot(input_size_adaptive_d[i], speed_adaptive_d[i], label=f"h_quater={i}")

ax[0][2].set_yscale("log")
ax[0][2].legend()

data_adaptive_a_contextual = [getBenchmark("BM_benchmark_adaptive_a_contextual_coder", i) for i in ["0", "1", "2", "3", "4"]]

input_size_adaptive_a_contextual = [data_adaptive_a_contextual[i]['input_size'] for i in range(5)]
real_time_adaptive_a_contextual = [data_adaptive_a_contextual[i]['real_time'] for i in range(5)]
speed_adaptive_a_contextual = [real_time_adaptive_a_contextual[i] / input_size_adaptive_a_contextual[i] for i in range(5)]

for i in range(5):
    ax[1][0].plot(input_size_adaptive_a_contextual[i], speed_adaptive_a_contextual[i], label=f"h_quater={i}")

ax[1][0].set_yscale("log")
ax[1][0].legend()

data_adaptive_d_contextual = [getBenchmark("BM_benchmark_adaptive_d_contextual_coder", i) for i in ["0", "1", "2", "3", "4"]]

input_size_adaptive_d_contextual = [data_adaptive_d_contextual[i]['input_size'] for i in range(5)]
real_time_adaptive_d_contextual = [data_adaptive_d_contextual[i]['real_time'] for i in range(5)]
speed_adaptive_d_contextual = [real_time_adaptive_d_contextual[i] / input_size_adaptive_d_contextual[i] for i in range(5)]

for i in range(5):
    ax[1][1].plot(input_size_adaptive_d_contextual[i], speed_adaptive_d_contextual[i], label=f"h_quater={i}")

ax[1][1].set_yscale("log")
ax[1][1].legend()

data_adaptive_a_contextual_improved = [getBenchmark("BM_benchmark_adaptive_a_contextual_improved_coder", i) for i in ["0", "1", "2", "3", "4"]]

input_size_adaptive_a_contextual_improved = [data_adaptive_a_contextual_improved[i]['input_size'] for i in range(5)]
real_time_adaptive_a_contextual_improved = [data_adaptive_a_contextual_improved[i]['real_time'] for i in range(5)]
speed_adaptive_a_contextual_improved = [real_time_adaptive_a_contextual_improved[i] / input_size_adaptive_a_contextual_improved[i] for i in range(5)]

for i in range(5):
    ax[1][2].plot(input_size_adaptive_a_contextual_improved[i], speed_adaptive_a_contextual_improved[i], label=f"h_quater={i}")

ax[1][2].set_yscale("log")
ax[1][2].legend()

data_adaptive_d_contextual_improved = [getBenchmark("BM_benchmark_adaptive_d_contextual_improved_coder", i) for i in ["0", "1", "2", "3", "4"]]

input_size_adaptive_d_contextual_improved = [data_adaptive_d_contextual_improved[i]['input_size'] for i in range(5)]
real_time_adaptive_d_contextual_improved = [data_adaptive_d_contextual_improved[i]['real_time'] for i in range(5)]
speed_adaptive_d_contextual_improved = [real_time_adaptive_d_contextual_improved[i] / input_size_adaptive_d_contextual_improved[i] for i in range(5)]

for i in range(5):
    ax[2][0].plot(input_size_adaptive_d_contextual_improved[i], speed_adaptive_d_contextual_improved[i], label=f"h_quater={i}")

ax[2][0].set_yscale("log")
ax[2][0].legend()

data_ppma = [getBenchmark("BM_benchmark_ppma_coder", i) for i in ["0", "1", "2", "3", "4"]]

input_size_ppma = [data_ppma[i]['input_size'] for i in range(5)]
real_time_ppma = [data_ppma[i]['real_time'] for i in range(5)]
speed_ppma = [real_time_ppma[i] / input_size_ppma[i] for i in range(5)]

for i in range(5):
    ax[2][1].plot(input_size_ppma[i], speed_ppma[i], label=f"h_quater={i}")

ax[2][1].set_yscale("log")
ax[2][1].legend()

data_ppmd = [getBenchmark("BM_benchmark_ppmd_coder", i) for i in ["0", "1", "2", "3", "4"]]

input_size_ppmd = [data_ppmd[i]['input_size'] for i in range(5)]
real_time_ppmd = [data_ppmd[i]['real_time'] for i in range(5)]
speed_ppmd = [real_time_ppmd[i] / input_size_ppmd[i] for i in range(5)]

for i in range(5):
    ax[2][2].plot(input_size_ppmd[i], speed_ppmd[i], label=f"h_quater={i}")

ax[2][2].set_yscale("log")
ax[2][2].legend()

plt.show()
```
