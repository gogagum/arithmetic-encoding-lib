```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = open("../build/benchmark/results.json")
data = json.load(file)
```

```python
benchmarks = pd.DataFrame(data['benchmarks'])
```

```python
benchmarks
```

```python
benchmarks[["coder", "input_size", "h_quater"]]\
  = benchmarks["name"].str.split('/', expand=True)
```

```python
benchmarks = benchmarks.astype({"input_size" : "int64"})
```

```python
benchmarks
```

```python
benchmarks.where(benchmarks.coder == "BM_benchmark_adaptive_coder").where(benchmarks.h_quater == "2").dropna()
```

```python
plt_data = dict()

plt_data[('adaptive', 2)] = benchmarks.where(benchmarks.coder == "BM_benchmark_adaptive_coder")\
                                      .where(benchmarks.h_quater == "2").dropna()
plt_data[('adaptive_d', 2)] = benchmarks.where(benchmarks.coder == "BM_benchmark_adaptive_d_coder")\
                                        .where(benchmarks.h_quater == "2").dropna()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

data_adaptive_2 = plt_data[('adaptive', 2)]
data_adaptive_a_2 = plt_data[('adaptive_d', 2)]

ax.plot(data_adaptive_2['input_size'], data_adaptive_2['real_time'] / data_adaptive_2['input_size'])
ax.plot(data_adaptive_a_2['input_size'], data_adaptive_a_2['real_time'] / data_adaptive_a_2['input_size'])


ax.set_yscale("log")

plt.show()
```
