```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = open("../build/benchmark/encode_results.json")
data = json.load(file)
```

```python
benchmarks = pd.DataFrame(data['benchmarks'])
```

```python
benchmarks
```

```python
benchmarks[["coder", "input_size", "h_quater"]]\
  = benchmarks["name"].str.split('/', expand=True)
```

```python
benchmarks = benchmarks.astype({"input_size" : "int64"})
```

```python
benchmarks
```

```python
set(benchmarks["coder"].to_list())
```

```python
def getBenchmark(coder_name, h_quater):
    return benchmarks.where(benchmarks.coder == coder_name).where(benchmarks.h_quater == h_quater).dropna()
```

```python
getBenchmark("BM_benchmbenchmarks = benchmarks.astype({"input_size" : "int64"})ark_adaptive_coder", "2")
```

```python
def drawEncodeBenchmark(ax, benchmark_name: str, title: str):
    data = [getBenchmark(benchmark_name, i) for i in ["0", "1", "2", "3", "4"]]

    input_size = [data[i]['input_size'] for i in range(5)]
    real_time = [data[i]['real_time'] for i in range(5)]
    speed = [real_time[i] / input_size[i] for i in range(5)]

    for i in range(5):
        ax.plot(input_size[i], speed[i], label=f"h_quater={i}")

    ax.set_title(title)
    ax.set_yscale("log")
    ax.legend()
```

```python
fig, ax = plt.subplots(5, 3, figsize=(15, 25))

drawEncodeBenchmark(ax[0][0], "BM_benchmark_adaptive_coder", "Adaptive coder")
drawEncodeBenchmark(ax[0][1], "BM_benchmark_adaptive_a_coder", "Adaptive A coder")
drawEncodeBenchmark(ax[0][2], "BM_benchmark_adaptive_d_coder", "Adaptive D coder")
drawEncodeBenchmark(ax[1][0], "BM_benchmark_adaptive_a_contextual_coder", "Adaptive A contextual coder")
drawEncodeBenchmark(ax[1][1], "BM_benchmark_adaptive_d_contextual_coder", "Adaptive D contextual coder")
drawEncodeBenchmark(ax[1][2],
                    "BM_benchmark_adaptive_a_contextual_improved_coder",
                    "Adaptive A improved contextual coder")
drawEncodeBenchmark(ax[2][0],
                    "BM_benchmark_adaptive_d_contextual_improved_coder",
                    "Adaptive D improved contextual coder")
drawEncodeBenchmark(ax[2][1], "BM_benchmark_ppma_coder", "PPMA coder")
drawEncodeBenchmark(ax[2][2], "BM_benchmark_ppmd_coder", "PPMD coder")
drawEncodeBenchmark(ax[3][0], "BM_benchmark_esc_ppma_coder", "Esc PPMA coder")
drawEncodeBenchmark(ax[3][1], "BM_benchmark_esc_ppmd_coder", "Esc PPMD coder")
drawEncodeBenchmark(ax[3][2], "BM_benchmark_esc_adaptive_a_coder", "Esc adaptive A coder")
drawEncodeBenchmark(ax[4][0], "BM_benchmark_esc_adaptive_d_coder", "Esc adaptive D coder")
ax[4][1].axis('off')
ax[4][2].axis('off')

plt.show()
```
